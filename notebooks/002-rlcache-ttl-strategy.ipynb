```python
config_file = '../configs/rl_ttl_strategy.json'
with open(config_file, 'r') as fp:
    CONFIG = json.load(fp)

exp_name = CONFIG['experiment_name']
results_dir = f'results/notebooks/{exp_name}'
cache = storage_from_config(CONFIG['cache_backend_settings'])
shared_cache_info = CacheInformation(cache.capacity(), size_check_func=cache.size)

caching_strategy , _, ttl_strategy = strategies_from_config(CONFIG['cache_manager_settings'], results_dir, shared_cache_info)
```
